```python
import matplotlib.pyplot as plt
```

```python
x = range(-50,50)
y = [i**2 - i + 2 for i in x]
```

## Graph for 
y =  x^2 - x +2 

```python
plt.plot(x,y)
plt.annotate('0,0', xy=(0, 0), xytext=(-10, 100),
             arrowprops={'facecolor': 'blue'})
```

```python
x0 = range(0,10)
y0 = [i**(1/2) for i in x0]
```

```python
import math 
x1 = range(10)
y1 = [i**3 for i in x1]
x2 = range(10)
y2 = [math.tan(i) for i in x2]
x3 = x2
y3 = [2**i for i in x3]
```

```python
plt.subplot(221)
plt.plot(x0,y0)
plt.subplot(222)
plt.plot(x1,y1)
plt.subplot(223)
plt.plot(x2,y2)
plt.subplot(224)
plt.plot(x3,y3)
```

```python
plt.plot(x0,y0, label='y = SQRT(x)')
plt.plot(x1,y1, label = 'y = x^3')
plt.plot(x2,y2, label = 'y = tan(x)')
plt.plot(x3,y3, label = 'y = 2^x')
plt.legend()
```
